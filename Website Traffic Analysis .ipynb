```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample website traffic data (can be replaced with your actual data)
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Visitors': [100, 150, 120, 200, 180],
    'Page_Views': [500, 600, 550, 700, 650]
}
```

```python
# Convert data into DataFrame
df = pd.DataFrame(data)
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set 'Date' column as index
df.set_index('Date', inplace=True)
```

```python
# Calculate total visitors and page views
total_visitors = df['Visitors'].sum()
total_page_views = df['Page_Views'].sum()
```

```python
# Print total visitors and page views
print("Total Visitors:", total_visitors)
print("Total Page Views:", total_page_views)
```

```python
# Plot website traffic
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Visitors'], marker='o', label='Visitors')
plt.plot(df.index, df['Page_Views'], marker='s', label='Page Views')
plt.title('Website Traffic Analysis')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
